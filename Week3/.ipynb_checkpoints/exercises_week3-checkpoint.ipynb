- `salary`: >50K,<=50K

```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Yorko/mlcourse.ai/master/data/adult.data.csv'
data = pd.read_csv(url)
data.head()
```

**1. How many men and women (*sex* feature) are represented in this dataset?** 

```python
print('There are', sum(data['sex'] == "Female"), 'Females')
print('There are', sum(data['sex'] == "Male"), 'Males')
```

**2. What is the average age (*age* feature) of women?**

```python
print('Mean age of females is', data[data['sex'] == "Female"].age.mean())
```

**3. What is the percentage of German citizens (*native-country* feature)?**

```python
print('There are', (sum(data['native-country'] == "Germany")/data['native-country'].count())*100, '% Germans')
```

**4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (*salary* feature) and those who earn less than 50K per year?**

```python
print('Salary <= 50K', data[data['salary'] == '<=50K'].age.std())
print('Salary <= 50K', data[data['salary'] == '>50K'].age.std())
```

**6. Is it true that people who earn more than 50K have at least high school education? (*education = HS-grad, Some-College, Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or *Doctorate*)**

```python
# You code here
education = ['HS-grad','Some-College','Bachelors','Prof-school','Assoc-acdm','Assoc-voc','Masters','Doctorate']
sum(data[data['salary'] == '>50K'].education.isin(education) == False)

print('No')
```

**7. Display age statistics for each race (*race* feature) and each gender (*sex* feature). Use *groupby()* and *describe()*. Find the maximum age of men of *Amer-Indian-Eskimo* race.**

```python
data.groupby(['race','sex']).describe() 
```

```python
print('Maximum age of Amer-Indian-Eskimo is 82')
```

**8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (*marital-status* feature)? Consider as married those who have a *marital-status* starting with *Married* (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.**

```python
data[(data['salary'] == '>50K') & (data['marital-status'].str.lower().str.startswith('married'))].age.count()/sum(data['salary'] == '>50K')*100

print('85.9%')
```

**9. What is the maximum number of hours a person works per week (*hours-per-week* feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?**

```python
data.groupby(['salary','hours-per-week']).describe()
```

```python
print('max hours per week', data['hours-per-week'].max())
print('Number that work for 99 hours week', sum(data['hours-per-week'] == 99)) 
print('Percentage that earns more than 50K', data[(data['salary'] == '>50K') & (data['hours-per-week'] == 99)].age.count()/sum(data['hours-per-week'] == 99)*100)
```

**10. Count the average time of work (*hours-per-week*) for those who earn a little and a lot (*salary*) for each country (*native-country*). What will these be for Japan?**

```python
data.groupby(['salary','hours-per-week','native-country']).describe()
```
